# 4.0) Limpeza de dados

  4.1) Removendo dados outliers e valores ausentes.

  Primeiro arrumei o dataframe utilizando a planilha google, assim consegui que todas as strings que nomeavam cada coluna pudesse estar realmente no topo da tabela. link para tabela usada: https://docs.google.com/spreadsheets/d/1-Sj-yjWQF6IL-uPh2mxM7Pupz16vpttaAHdq6xtq_9o/edit?usp=sharing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('kc_house_data1.csv')
data.drop(['waterfront','view','sqft_above','sqft_basement','yr_built','yr_renovated','sqft_living15','sqft_lot15'], inplace=True, axis=1)
data
```

O objetivo dessa limpeza de colunas é diminuir a quantidade de dados que não serâo relevantes para nossa análise ou são muito imprecisos.

Não vemos de primeiro momento nenhuma relação desses dados com o valor do imóvel.

```python
data[data.columns[2]] = data[data.columns[2]].replace('[\$,]', '', regex=True).astype(float)

data[data.columns[2:10]] = data[data.columns[2:10]].astype('int')
data
```

```python
data = data.dropna(axis = 0, how ='any')
data
```

Dessa forma garantimos que não tem nenhum valor nulo dentro desse data set.

# 5.0)Exploração de dados

### 5.1) Visualização dos dados

```python
apt_vendidos=data[data.groupby('id')[data.columns[2]].transform('nunique') > 1]
apt_vendidos
apt_vendidos[apt_vendidos.columns[0]] = apt_vendidos[apt_vendidos.columns[0]]/10000000
apt_vendidos
```

Se nós analisarmos os apt que ja foram vendidos a 
gente consegue achar um padrão para poder analisar o que é relevante para ele ter uma valorização ou uma desvalorização.

```python
apt_vendidos['diff'] = apt_vendidos.groupby('id')[apt_vendidos.columns[2]].shift(-1).sub(apt_vendidos[apt_vendidos.columns[2]])
apt_vendidos
```

Com esse código conseguimos ver a diferença de valor de revenda e ver qual o padrão estabelecido.

```python
import plotly.express as px
fig = px.scatter(apt_vendidos, x="id", y="diff")
fig.show()
```

```python
prejuizo = apt_vendidos[apt_vendidos['diff'] < 0]
prejuizo
```

Podemos notar alguns padrões para o investimento ter prejuízo, sentimos um padrão no bedrooms ou 2 ou 4, também notamos que a condition está todas no valor 3, além disso notamos que pelo menos o tamanho do terreno é pelo menos 3x o tamanho da casa em 4 de 6 imóveis.

```python
filter_values = np.where((data[data.columns[3]] == 4) & (data[data.columns[8]] == 3) & (data['sqft_lot'] >= (4*data['sqft_living'])))
low5imoveis = data.loc[filter_values]
low5imoveis
```

```python
lucro = apt_vendidos[apt_vendidos['diff'] > 0]
lucro = lucro.sort_values(by="diff",ascending=False)
lucro.head()
```

Nos gráficos acima temos os imóveis que tiveram prejuizo e no outro temos os imóveis que tiveram mais lucro na revenda.
Notamos que a maior taxa de lucratividade está em imóveis simples, com apenas 1 piso, com apenas 1 banheiro e de 2 a 3 quartos e também vemos que a localidade faz diferença na valorização dos imóveis quando vemos que os 5 imóveis que mais valorizaram tem apenas 3 zipcode.

```python
filtered_values = np.where((data[data.columns[7]] <= 1) & (data[data.columns[4]] <=1)&(data['bedrooms'] <=3)&(data['bedrooms'] >=2)&(data['zipcode'] == 98117))
top5imoveis = data.loc[filtered_values]
top5imoveis
```

Como o zipcode com maior valorização é o 98117, utilizamos ele como um local exemplo de valorização de imóveis.
Na tabela acima os imóveis tem as características ja citadas no ultimo passo que foi encontrar o padrão das casas com maior lucratividade na revenda. 
No caso para lucrarmos mais, temos que investir mais, portanto podemos ordenar a tabela acima pelo preço e investir nos 5 imóveis mais caros porque provavelmente vamos ter uma lucratividade boa na revenda.
